```python
import pandas as pd
import sklearn
import numpy as np

```

```python
#1 load the dataset
dataset = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")
dataset.head(5)
```

```python
#2 Dimensions 
dataset.shape
```

```python
# Describe the dataset
dataset.describe()
```

```python
#3 Check the variable (Numberical)
numerical_variable = dataset.columns[dataset.dtypes != 'object']
dataset[numerical_variable]
```

```python
#3 Check the variable categorical 
categorical_variable = dataset.columns[dataset.dtypes == 'object']
dataset[categorical_variable]
```

```python
#4 Missing Value 
dataset.isnull()
```

```python
#5 Find out the the percentage of missing values
dataset.isnull().sum()/len(dataset)
```

```python
# Numerical data missing percentage
dataset[numerical_variable].isnull().sum()/len(dataset)
```

```python
# Categorical data missing percentage
dataset[categorical_variable].isnull().sum()/len(dataset)
```

```python
#6 Replace
dataset['Parch'].replace(0,1)
```

```python
#7 Fill up missing value 
dataset.fillna('C148')
```

```python
#8 Create the rows and add it with the original dataset 
d = {'PassengerId':1,'Survived': 0, 'Pclass':0,'Name':'Testing','Sex':'Male','Age':23.0,'SibSp':1,'Parch':0,'Ticket':2212,'Fare':2322,'Cabin':'C2422','Embarked':'S'}
row_insert = pd.Series(data=d)
dataset.append(row_insert,ignore_index=True)
```

```python
# 9 Choose exact rows from the dataset
dataset.iloc[2:10]
```

```python
# 10 Rename the column
dataset.rename(columns = {'Name':'Full Name'})
```

```python
#Find the min, max, sum, count
print(dataset.sum())
print(dataset.min())
print(dataset.max())
print(dataset.count())
```

```python
# 12 Delete columns
dataset.drop('Age',axis=1).head(5)
```

```python
#12 Delete rows 
dataset.drop([1],axis=0)
```
